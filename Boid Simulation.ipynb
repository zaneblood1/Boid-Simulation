```python
#2D boid simulation
import random

Vmax=10
Amax=10

Lx=100
Ly=100

#Maximum radius a Bird can see
radius=100

#How close a bird can get to another before it wants to steer away
tooClose=5

#Factors indicating the strength of Alignment, Cohesion, and Separation
Alignment=2
Cohesion=1
Separation=10

#Define a bird object that has 2D position, velocity, and acceleration
class Bird:
    def __init__(self, x, y, vx, vy, ax, ay):
        self.x=x
        self.y=y
        
        self.vx=vx
        self.vy=vy
        
        self.ax=ax
        self.ay=ay
        
    
#Number of birds
numBirds=100

#Number of time steps
numTstep=100

#Time step
dt=1

#List of birds
Birds=[]

#frames to animate
frames=[]

#Initializing the birds
for i in range(numBirds):

    Birds.append(Bird(random.random()*Lx, random.random()*Ly, random.uniform(-Vmax,Vmax), random.uniform(-Vmax,Vmax), random.uniform(0,Amax), random.uniform(0,Amax)))

#Frames that hold the x and y position vectors of each bird in the flock at a given time
xFrames=[]
yFrames=[]
   
#The x and y position vectors of the entire flock at a given time
xvec=[]
yvec=[]

#Loop through the Birds list for the first initialization at t=0 and update xvec, yvec, xFrames, and yFrames
for i in range(len(Birds)):  
    
    xvec.append(Birds[i].x)
    yvec.append(Birds[i].y)
    
xFrames.append(xvec)
yFrames.append(yvec)

#Loop to evolve the system
for t in range(numTstep):
    
    #Position vectors of birds at a given time step
    xvec=[]
    yvec=[]
    
    #Loop through the birds
    for i in range(numBirds):
        
        #Alignment variables
        vxsum=0
        vysum=0
        numsum=0
        
        #Cohesion variables
        xsum=0
        ysum=0
        
        #Separation variables
        xclosesum=0
        yclosesum=0
        numclosesum=0
        
        #Loop through the birds list again comparing each bird to the current bird we are looking at in the 
        #outermost bird loops
        for j in range(numBirds):
            
            #Update the alignment variables
            if j!=i and (Birds[i].x-Birds[j].x)**2+(Birds[i].y-Birds[j].y)**2<=radius**2:
                vxsum=Birds[j].vx+vxsum
                vysum=Birds[j].vy+vysum
                numsum=numsum+1
                
            #calculating the center of mass / updating cohesion variables
            xsum=xsum+Birds[j].x
            ysum=ysum+Birds[j].y
            
            #Updating the separation variables
            if j!=i and (Birds[i].x-Birds[j].x)**2+(Birds[i].y-Birds[j].y)**2<=tooClose**2:
                xclosesum=Birds[j].x+xclosesum
                yclosesum=Birds[j].y+yclosesum
                numclosesum=numclosesum+1
             
        #Calculating COM for cohesion
        xcom=xsum/numBirds
        ycom=ysum/numBirds
        
        #Updating variables for separation (steering away from a local COM...)
        xclosecom=0
        yclosecom=0
    
        if numclosesum>0:
            
            xclosecom=xclosesum/numclosesum
            yclosecom=yclosesum/numclosesum
        
        #Update the velocity for alignment
        if numsum>=1:
        
            Birds[i].vx=(vxsum/numsum)*Alignment
            Birds[i].vy=(vysum/numsum)*Alignment
            
        else:
            
            Birds[i].vx=Birds[i].vx
            Birds[i].vy=Birds[i].vy
            
        #Align velocity towards center of mass
        Birds[i].vx=Birds[i].vx+(xcom-Birds[i].x)*Cohesion
        Birds[i].vy=Birds[i].vy+(ycom-Birds[i].y)*Cohesion
        
        #Steer to avoid other boids
        if xclosecom>0:
            Birds[i].vx=Birds[i].vx+(-xclosecom+Birds[i].x)*Separation
        if yclosecom>0:
            Birds[i].vy=Birds[i].vy+(-yclosecom+Birds[i].y)*Separation
        
        #Update the position
        Birds[i].x=Birds[i].vx*dt+Birds[i].x
        Birds[i].y=Birds[i].vy*dt+Birds[i].y
        
        #Make the birds bound within the screen
        if Birds[i].x>=Lx or Birds[i].x<=0:
            Birds[i].vx=-Birds[i].vx
            
        if Birds[i].y>=Ly or Birds[i].y<=0:
            Birds[i].vy=-Birds[i].vy
            
        #Make sure the velocity is set to a maximum
        if Birds[i].vx>Vmax:
            Birds[i].vx=Vmax
        if Birds[i].vx<-Vmax:
            Birds[i].vx=-Vmax
        if Birds[i].vy>Vmax:
            Birds[i].vy=Vmax
        if Birds[i].vy<-Vmax:
            Birds[i].vy=-Vmax
        
        #Add the positions to a vector
        xvec.append(Birds[i].x)
        yvec.append(Birds[i].y)
        
    #Add the position vectors to frames vectors for animation
    xFrames.append(xvec)
    yFrames.append(yvec)
```

```python
import matplotlib.pyplot as plt

#Animation cell 
%matplotlib

truth=True

plt.clf()

dD=10

for i in range(len(xFrames)):

        plt.scatter(xFrames[i], yFrames[i], s=2, color='w')
        ax = plt.gca()
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.title("Birds")
        plt.xlim(0-dD,Lx+dD)
        plt.ylim(0-dD,Ly+dD)
#         plt.xlim(-1000,1000)
#         plt.ylim(-1000,1000)
        ax.set_facecolor('k')
        plt.show()
        plt.pause(0.1)

        if i<len(xFrames)-1:
            plt.clf()

```
